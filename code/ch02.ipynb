## 4. 맷플롯립으로 그래프 그리기
: matplotlib은 파이썬 과학 생태계의 표준 그래프 패키지이다. 

```
import matplotlib.pyplot as plt
```

1. 선 그래프 그리기
: plt.plot(x축의 값, y축의 값)

```
plt.plot([1,2,3,4,5],[1,4,9,16,25])
plt.show()
```

2. 산점도 그리기
: scatter point란 x축, y축 값을 이용하여 점으로 그래프를 그린 것

```
plt.scatter([1,2,3,4,5],[1,4,9,16,25])
plt.show()
```

3. 넘파이 배열로 산점도 그리기
: random.rnadn()함수를 사용하여 표준 정규 분포를 따르는 난수를 만들 수 있음

```
x = np.random.randn(1000) # 표준 정규 분포를 따르는 난수 1000개를 만듦
y = np.random.randn(1000)
plt.scatter(x,y)
plt.show()
```
